#### Graph Functions
1. plot                         - Simple Line Plot
2. plot.area                    - Filled Version of Line Plot  
3. plot.bar   / plot.barh       - Bar Graph
4. plot.box                     - Box Plot
5. plot.density                 - Density Line
6. plot.hexbin                  - Hexagon Version of Scatter with Density
7. plot.hist                    - Histogram
8. plot.kde                     - Kernel Density Estimator (Density Graph)
9. plot.pie                     - Pie Chart of Categorical Data
10. plot.scatter                - Scatter Graph

```python
df2.plot();plt.title('LinePlot');plt.show()
df2.plot.area();plt.title('Area Plot');plt.show()
df2.plot.bar();plt.title('Bar Plot');plt.show()
df2.plot.box();plt.title('Box Plot');plt.show()
df2.plot.barh();plt.title('Bar Horizontal Plot');plt.show()
df2.plot.density();plt.title('Density Plot');plt.show()
df2.plot.kde();plt.title('KDE Plot');plt.show()
df2.plot.hist();plt.title('Histogram Plot');plt.show()

df2.plot.scatter(x='a',y='b');plt.title('Scatter Plot');plt.show()

df2.plot.hexbin(x='a',y='b',C='c')
plt.title('Hexbin Plot');plt.show()
```

```python
import seaborn as sns
```

###### Categories
1. Categorical Plots
    a. countplot
    b. barplot (one categorical,one numerical)
    c. boxplot
    d. voilinplot
    e. swarmplot
2. Distribution Plots
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix Plots
    a. heatmap
    b. clustermap
    c. pairplot (same as above)
4. Regression PLots
    a. lmplot
    b.regplot
5. Grid Plots
    a. PairGrid
    b. JointGrid

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe(include='all')
```

```python
tips.value_counts('day')
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=max)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=min)
```

```python
from statistics import mode
sns.barplot(x='day',y='total_bill',data=tips,estimator=mode)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.sum)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
plt.show()
sns.violinplot(x='day',y='total_bill',data=tips,hue='time')
plt.show()
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
plt.show()
```
